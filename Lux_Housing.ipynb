```python
lxdf[['Connectivity_Score', 'Amenity_Score', 'Locality_Infra_Score']] = (lxdf[['Connectivity_Score', 'Amenity_Score', 'Locality_Infra_Score']].astype('float64').round(2))
```

```python
lxdf['Ticket_Price_Cr'] = (lxdf['Ticket_Price_Cr'].astype(str).str.replace('₹','', regex=False)
                            .str.replace('Cr','', regex=False).str.strip().astype(float).round(2))
```

```python
# Make all text uppercase and remove extra spaces
lxdf['Configuration'] = lxdf['Configuration'].str.strip().str.upper()

# Optional: Add space between number and BHK (e.g., "3BHK" → "3 BHK")
# "Take whatever digit I matched in the first parentheses, then add a space, then add the letters BHK."
lxdf['Configuration'] = lxdf['Configuration'].str.replace(r'(\d)(BHK)', r'\1 BHK', regex=True)

# Check the cleaned counts
print(lxdf['Configuration'].value_counts())
```

```python
lxdf['Micro_Market'] = lxdf['Micro_Market'].str.strip().str.title()
lxdf['Micro_Market'].value_counts()
```

```python
lxdf['Unit_Size_Sqft'] = pd.to_numeric(lxdf['Unit_Size_Sqft'], errors='coerce')

lxdf.loc[lxdf['Unit_Size_Sqft'] < 0,'Unit_Size_Sqft']= np.nan

lxdf['Unit_Size_Sqft'] = lxdf['Unit_Size_Sqft'].fillna(lxdf['Unit_Size_Sqft'].median())
```

```python
# Convert to float (invalid parsing becomes NaN)
lxdf['Ticket_Price_Cr'] = pd.to_numeric(lxdf['Ticket_Price_Cr'], errors='coerce')

# Replace negative values with NaN
lxdf.loc[lxdf['Ticket_Price_Cr'] < 0, 'Ticket_Price_Cr'] = np.nan

# Calculate median of valid values
median_value = lxdf['Ticket_Price_Cr'].median()

# Fill NaN with median
lxdf['Ticket_Price_Cr'] = lxdf['Ticket_Price_Cr'].fillna(median_value)
```

```python
lxdf['Buyer_Comments'] = lxdf['Buyer_Comments'].fillna('No Comments Provided.')
```

```python
lxdf['Buyer_Comments'].unique()
```

```python
lxdf['Quarter_Number'] = pd.PeriodIndex(lxdf['Purchase_Quarter'], freq='Q').quarter
```

```python
num_cols = ['Unit_Size_Sqft', 'Ticket_Price_Cr', 'Amenity_Score']

for col in num_cols:
    print(f"\nColumn: {col}")
    print("Skewness:", lxdf[col].skew())
    print(lxdf[col].describe())
```

```python
lxdf['Amenity_Score'] = (lxdf['Amenity_Score']).fillna(lxdf['Amenity_Score'].mean())
lxdf['Amenity_Score']
```

```python
lxdf['Ticket_Price_Cr'].isnull().sum()
```

```python
lxdf['Price_per_Sqft'] = ((lxdf['Ticket_Price_Cr']*1e7)/lxdf['Unit_Size_Sqft']).round(2)
lxdf['Price_per_Sqft']
```

```python
lxdf['booking_flag'] = lxdf['Purchase_Quarter'].apply(lambda x: 0 if pd.isna(x) or x.strip() == '' else 1)
lxdf
```

```python
lxdf['Purchase_Quarter'].unique()
```

```python
lxdf.isnull().sum()
```

```python
lxdf.columns = (
    lxdf.columns.str.strip()        # remove leading/trailing spaces
              .str.lower()        # lowercase
              .str.replace(' ', '_')  # replace spaces with _
              .str.replace('-', '_')  # replace hyphens with underscores
)
```

```python
q1 = '''SELECT sum(ticket_price_cr) from luxury_house;'''
count = pd.read_sql(q1,engine)
count
```

```python
q2 = '''SELECT developer_name,ROUND((sum(ticket_price_cr)::numeric),2) as totalprice_cr FROM luxury_house
GROUP BY developer_name
ORDER BY totalprice_cr desc
LIMIT 5'''
total = pd.read_sql(q2,engine)
total
```

```python
q3 = """SELECT developer_name,ROUND((AVG(ticket_price_cr)::numeric),2) AS avg_ticket_price_cr from luxury_house
GROUP BY developer_name
ORDER BY avg_ticket_price_cr desc;"""
avg_tp = pd.read_sql(q3,engine)
avg_tp
```

```python
q4 = '''SELECT developer_name,count(developer_name) as total_count FROM luxury_house
group by developer_name
order by total_count desc'''
tcount = pd.read_sql(q4,engine)
tcount
```

```python
q5 = '''SELECT developer_name,buyer_type,count(buyer_type) as total_count FROM luxury_house
group by buyer_type,developer_name
order by buyer_type'''
tbuyer = pd.read_sql(q5,engine)
tbuyer
```

```python
q9 = '''SELECT developer_name,sales_channel,COUNT(sales_channel) from luxury_house 
GROUP BY developer_name,sales_channel
ORDER BY COUNT(sales_channel) desc'''
count = pd.read_sql(q9,engine)
count
```

```python
q10 = '''SELECT micro_market,count(micro_market) as totalmarket_count FROM luxury_house
group by micro_market
order by totalmarket_count desc'''
mtcount = pd.read_sql(q10,engine)
mtcount
```

```python
#1. Market Trends: bookings per quarter and micro market
q11 = '''SELECT purchase_quarter, quarter_number, micro_market, COUNT(*) AS booking_count
FROM luxury_house
GROUP BY purchase_quarter, quarter_number, micro_market
ORDER BY quarter_number, micro_market'''
mt = pd.read_sql(q11,engine)
mt
```

```python
#c2. Builder Performance
q12 = '''SELECT developer_name,
       SUM(ticket_price_cr) AS total_ticket_sales,
       AVG(ticket_price_cr) AS avg_ticket_size
FROM luxury_house
GROUP BY developer_name
ORDER BY total_ticket_sales DESC;'''
bp = pd.read_sql(q12,engine)
bp
```

```python
#3. Amenity Impact (conversion proxy = share of primary transactions if applicable)
q13 = '''SELECT amenity_score,
       COUNT(DISTINCT project_name) AS project_count,
       1.0 * SUM(CASE WHEN transaction_type = 'Primary' THEN 1 ELSE 0 END) / COUNT(*) AS primary_share
FROM luxury_house
GROUP BY amenity_score
ORDER BY amenity_score;'''
ai = pd.read_sql(q13,engine)
ai
```

```python
# 4. Booking Conversion by micro market (proxy using transaction_type distribution)
q14 ='''SELECT  transaction_type, COUNT(*) AS cnt
FROM luxury_house
GROUP BY  transaction_type;'''
bc =pd.read_sql(q14,engine)
bc
```

```python
#5. Configuration Demand
q15 = '''SELECT configuration, COUNT(*) AS booking_count
FROM luxury_house
GROUP BY configuration
ORDER BY booking_count DESC;'''
cd = pd.read_sql(q15,engine)
cd
```

```python
# 6. Sales Channel Efficiency (distribution by transaction_type)
q16 = '''SELECT sales_channel, transaction_type, COUNT(*) AS cnt
FROM luxury_house
GROUP BY sales_channel, transaction_type;'''
sc = pd.read_sql(q16,engine)
sc
```

```python
# 7. Quarterly Builder Contribution
q17 = '''SELECT developer_name, purchase_quarter, quarter_number,
       SUM(ticket_price_cr) AS total_ticket_sales
FROM luxury_house
GROUP BY developer_name, purchase_quarter, quarter_number
ORDER BY developer_name, quarter_number;'''
qb = pd.read_sql(q17,engine)
qb
```

```python
#-- 8. Possession Status vs Buyer Type
q18 ='''SELECT possession_status, buyer_type, COUNT(*) AS cnt
FROM luxury_house
GROUP BY possession_status, buyer_type
ORDER BY possession_status;'''
pb = pd.read_sql(q18,engine)
pb
```

```python
#- 9. Geographical Insights
q19 = '''SELECT micro_market, COUNT(*) AS project_count
FROM  luxury_house
GROUP BY micro_market
ORDER BY project_count DESC;'''
gi = pd.read_sql(q19,engine)
gi
```

```python
#-- 10. Top Performers (Top 5 by revenue)
q20 = '''SELECT developer_name,
       SUM(ticket_price_cr) AS total_revenue,
       1.0 * SUM(CASE WHEN transaction_type = 'Primary' THEN 1 ELSE 0 END) / COUNT(*) AS primary_share
FROM luxury_house
GROUP BY developer_name
ORDER BY total_revenue DESC
LIMIT 5;'''
tp = pd.read_sql(q20,engine)
tp
```

```python
q11 = '''SELECT micro_market,
       MAX(price_per_sqft) AS max_price_per_sqft,
       MIN(price_per_sqft) AS min_price_per_sqft,
       AVG(price_per_sqft) AS avg_price_per_sqft
FROM luxury_house
GROUP BY micro_market
ORDER BY avg_price_per_sqft DESC;'''
mma = pd.read_sql(q11,engine)
mma
```
